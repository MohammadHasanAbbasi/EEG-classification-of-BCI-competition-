```python
import matplotlib.pyplot as plt
import numpy as np

def plot_signal(X,label,channel_num=22):
    for i in range(channel_num):
        plt.figure(figsize=(12,4))
        plt.xlabel('timesteps')
        plt.title('{}: channel {} '.format(label,i+1))
        plt.plot(X[i])
        plt.show()
```

```python
with open('integrated_data.npy', 'rb') as f:
    X = np.load(f)
    y = np.load(f)
```

```python
event_codes={769:'Left Hand',770:'Right Hand',771:'Foot',772:'Tongue'}

for index in event_codes.keys():
    label_index = (np.where(y==index)[0][0])
    plot_signal(X[label_index],event_codes[index],channel_num=1)
```
